```python

df1_scaled = z.fit_transform(dataset1)
df2_scaled = z.fit_transform(dataset2)
df3_scaled = z.fit_transform(dataset3)
df4_scaled = z.fit_transform(dataset4)
df5_scaled = z.fit_transform(dataset5)
df6_scaled = z.fit_transform(dataset6[["x","y"]])
df7_scaled = z.fit_transform(dataset7)
df8_scaled = z.fit_transform(dataset8)
df9_scaled = z.fit_transform(dataset9)
```

# Use K-means, Agglomerative, DBSCAN

```python
from sklearn.cluster import KMeans
#kmeans = KMeans(n_clusters=2, random_state=0).fit(X)

from sklearn.cluster import AgglomerativeClustering
#agg = AgglomerativeClustering(n_clusters=4).fit(df_scaled[col_names])

from sklearn.cluster import DBSCAN
#dbscan = DBSCAN(eps=3, min_samples=2).fit(X)
```

```python
kmeans1 = KMeans(n_clusters=2).fit(df1_scaled)
kmeans2 = KMeans(n_clusters=2).fit(df2_scaled)
kmeans3 = KMeans(n_clusters=3).fit(df3_scaled)
kmeans4 = KMeans(n_clusters=2).fit(df4_scaled)
kmeans5 = KMeans(n_clusters=2).fit(df5_scaled)
kmeans6 = KMeans(n_clusters=3).fit(df6_scaled)
kmeans7 = KMeans(n_clusters=3).fit(df7_scaled)
kmeans8 = KMeans(n_clusters=2).fit(df8_scaled)
kmeans9 = KMeans(n_clusters=2).fit(df9_scaled)
```

```python
agg1 = AgglomerativeClustering(n_clusters=2).fit(df1_scaled)
agg2 = AgglomerativeClustering(n_clusters=2).fit(df2_scaled)
agg3 = AgglomerativeClustering(n_clusters=3).fit(df3_scaled)
agg4 = AgglomerativeClustering(n_clusters=2).fit(df4_scaled)
agg5 = AgglomerativeClustering(n_clusters=2).fit(df5_scaled)
agg6 = AgglomerativeClustering(n_clusters=3).fit(df6_scaled)
agg7 = AgglomerativeClustering(n_clusters=3).fit(df7_scaled)
agg8 = AgglomerativeClustering(n_clusters=2).fit(df8_scaled)
agg9 = AgglomerativeClustering(n_clusters=2).fit(df9_scaled)
```

```python
dbscan1 = DBSCAN(eps=0.2, min_samples=10).fit(df1_scaled)
dbscan2 = DBSCAN(eps=0.2, min_samples=10).fit(df2_scaled)
dbscan3 = DBSCAN(eps=0.5, min_samples=10).fit(df3_scaled)
dbscan4 = DBSCAN(eps=0.3, min_samples=5).fit(df4_scaled)
dbscan5 = DBSCAN(eps=0.5, min_samples=10).fit(df5_scaled)
dbscan6 = DBSCAN(eps=0.3, min_samples=10).fit(df6_scaled)
dbscan7 = DBSCAN(eps=0.5, min_samples=10).fit(df7_scaled)
dbscan8 = DBSCAN(eps=0.5, min_samples=10).fit(df8_scaled)
dbscan9 = DBSCAN(eps=0.3, min_samples=6).fit(df9_scaled)
```
