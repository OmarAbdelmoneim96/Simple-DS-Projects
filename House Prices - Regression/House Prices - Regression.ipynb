# Encoding

```python
df.dtypes
```

#### Split df Into Numerical & Categorical 

```python
df_categorical = df.select_dtypes(include=['object']).copy() #columns with type "Object"
df_categorical.head()
```

```python
quantitative = [f for f in df.columns if df.dtypes[f] != 'object']
df_numerical = df[quantitative]
df_numerical
```

```python
cleanup_nums = {"Street":     {"Grvl": 0, "Pave": 1},
                "LotShape":   {"IR3": 0, "IR2": 1, "IR1": 2, "Reg": 3},
                "LandContour":{"Low": 0, "HLS":1, "Bnk": 2, "Lvl": 3},
                "Utilities":  {"ELO": 0, "NoSeWa":1, "NoSewr": 2, "AllPub": 3},
                "LandSlope":  {"Sev": 0, "Mod":1, "Gtl": 2}, 
                "ExterQual":  {"Po": 0, "Fa":1, "TA": 2, "Gd": 3, "Ex": 4},
                "ExterCond":  {"Po": 0, "Fa":1, "TA": 2, "Gd": 3, "Ex": 4},
                "BsmtQual":   {"None":0, "Po": 1, "Fa":2, "TA": 3, "Gd": 4, "Ex": 5},
                "Bsmtcond":   {"None":0, "Po": 1, "Fa":2, "TA": 3, "Gd": 4, "Ex": 5},
                "BsmtExposure":   {"None":0, "No": 1, "Mn":2, "Av": 3, "Gd": 4},
                "BsmtFinType1":   {"None":0, "Unf": 1, "LwQ":2, "Rec": 3, "BLQ": 4, "ALQ": 5, "GLQ": 6},
                "BsmtFinType2":   {"None":0, "Unf": 1, "LwQ":2, "Rec": 3, "BLQ": 4, "ALQ": 5, "GLQ": 6},
                "HeatingQC":  {"Po": 0, "Fa":1, "TA": 2, "Gd": 3, "Ex": 4},
                "CentralAir":  {"N": 0, "Y":1},
                "Electrical":  {"Mix": 0, "FuseP":1, "FuseF": 2, "FuseA": 3, "SBrkr": 4},
                "KitchenQual":  {"Po": 0, "Fa":1, "TA": 2, "Gd": 3, "Ex": 4},
                "Functional":  {"Sal": 0, "Sev":1, "Maj2": 2, "Maj1": 3, "Mod": 4, "Min2": 5, "Min1": 6, "Typ": 7},
                "FireplaceQu":  {"None":0, "Po": 1, "Fa":2, "TA": 3, "Gd": 4, "Ex": 5},
                "GarageFinish": {"None": 0, "Unf":1, "RFn": 2, "Fin": 3},
                "GarageQual":  {"None":0, "Po": 1, "Fa":2, "TA": 3, "Gd": 4, "Ex": 5},
                "GarageCond":  {"None":0, "Po": 1, "Fa":2, "TA": 3, "Gd": 4, "Ex": 5},
                "PoolQC":  {"None":0, "Po": 1, "Fa":2, "TA": 3, "Gd": 4, "Ex": 5},
                "Fence":  {"None":0, "MnWw": 1, "GdWo":2, "MnPrv": 3, "GdPrv": 4},
                "PavedDrive":  {"N": 0, "P":1, "Y": 2}, 
                
                
               }


df_categorical = df_categorical.replace(cleanup_nums)
df_categorical.head()
```

#### Encoding Nominal Categorical Data Using "One Hot Encoding"

```python
df_categorical_nominal = df_categorical.select_dtypes(include=['object']).copy()
df_categorical_nominal
```

```python
df_categorical_ordinal = df_categorical.drop(columns=list(df_categorical_nominal.columns))
```

```python
OHE_results = pd.get_dummies(df_categorical_nominal, columns=list(df_categorical_nominal.columns)) 
```

```python

OHE_results.drop(columns=['MasVnrType_None',"GarageType_None",'BsmtCond_None'],inplace=True)
```

```python
OHE_results
```

```python
df_categorical_ordinal
```

#### Concatenation After Encoding

```python
df_encoded = pd.concat([df_numerical,OHE_results,df_categorical_ordinal], axis=1)
df_encoded
```

# Model

#### split to original train-test

```python
train = df_encoded[0:1460]
train
```

```python
x_train = train.drop(columns=['Id'])
x_train
```

```python
y_train
```

```python
test = df_encoded[1460:]
test
```

```python
x_test = test.drop(columns=['Id'])
x_test
```

#### ML Model

```python
lin_reg = LinearRegression().fit(x_train, y_train)
lin_reg.coef_
```

```python
y_pred = lin_reg.predict(x_train)
r2 = r2_score(y_train,y_pred)
print(r2)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor().fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_train)
r2 = r2_score(y_train,y_pred)
print(r2)
```
